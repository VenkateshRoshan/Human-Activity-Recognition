```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers , models
import os
import cv2
from matplotlib import pyplot as plt
```

```python
model = models.load_model('Human_Activity.h5')
selected = ['smoking' , 'running']
def predictLabel(img,size=224) :
    img = cv2.resize(img,(size,size))
    tst = []
    tst.append(img)
    tst = np.array(tst)/255.
    return selected[model.predict_classes(tst)[0][0]]
```

```python
img = cv2.imread('smoking_1.jpg')
plt.imshow(img)
plt.show()
predictLabel(img)
```
